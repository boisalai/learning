# Anaconda

## Install Anaconda on M1 Mac

See https://www.anaconda.com/products/distribution#Downloads[Anaconda Installers,window="_blank"].

See also https://pytorch.org/get-started/locally/[Get started - Start locally,window="_blank"].


## Commands 

Use `conda update anaconda` to update Anaconda.

Check version with `conda --version`.

To uninstall Anaconda: 

```bash
rm -rf ~/anaconda2
rm -rf ~/anaconda3
rm -rf ~/opt/anaconda3
```

List environments currently installed on my computer with `conda env list`.

Create a new environment called `ml`: `conda create --name ml python=3.7`.

Activate an environment: `conda activate ml`.

List all packages currently installed in the active environment: `pip freeze`.

To save the requirements list to a `requirements.txt` file: `pip freeze > requirements.txt`.

To a `environment.yml` file: `conda env export > environment.yml`.

Install packages from a `requirements.txt` file: `pip install -r requirements.txt`.

From a  `environment.yml` file: `conda env update --file environment.yml`.
