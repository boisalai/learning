```python
# See https://twitter.com/mervenoyann/status/1518563617209212932?s=20&t=XmuU1mFDk8u6qdmTNWkcvA
from transformers import pipeline

context = "Adventurer is approached by a mysterious stranger in the tavern for a new quest. He asks to hear if he would be able to help to locate an ancient relic."
question = "What is the purpose of the quest?"

QA_model = pipeline("question-answering")

answer = QA_model(question = question, context = context)

print(answer)
```
